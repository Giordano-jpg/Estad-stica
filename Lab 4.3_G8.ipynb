### Ejercicio 1 

Responder a las siguientes preguntas teóricas con <u>vuestras propias palabras</u>:

1. ¿Cuáles son las principales características de un modelo de árboles de decisión?
2. ¿Cuáles son las principales características de un modelo de support vector machine (SVM)?
3. ¿Cuáles son las principales diferencias entre un modelo de árbol de decisión y máquina de soporte vectorial?

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 01</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

Las características principales de los árboles de decisión son que se separaran los datos en distintos espacios. Cada nodo del árbol representa las caracterísitcas por las que se ha dividido el conjunto de datos, teniendo en cuenta la variable objetivo. Es útil para el ámbito financiero.
Los árboles utilizan un procso iterativo y que desciende que va formando grupos homogéneos.

Lo importante del árbol de decisión es su **interpretabilidad** porque generan reglas de decisión claras. Es **bueno para manejar datos heterogéneos**, ya que en un mismo conjunto de datos puede haber grandes variaciones. Puede haber un **sobreajuste** porque puede haber datos atípicos o muy diferentes a lo esperado. COn los árboles de decisión es **fácil evaluar** su resultado, ya que se toman las decisiones rapidamente. Aunque pueden ser **sensibles a cambios en los datos** lo que puede modificar la estructura del árbol.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

El modelo *support vector machine*, busca como objetivo optimizar un **hiperplano** (subespacio) que separa las clases con el máximo márgen posible. Se hace la separación usando la distancia entre el hiperplano y los puntos próximos de cada clase, siendo equidistante. La dimensión del hiperplano es "n-1", "n" es la cantidad de variables. Este modelo puede usar una función kernel para convertir os datos a un espacio con mayor dimensión donde se pueda hacer la separación lineal.

Resumidamente, es **apto para clasificación binaria**, además está diseñado para estos. Es **sensible a la escala de datos**, lo que requiere una normalización cuidadosa de los datos. Es eficaz generalizando datos no vistos, tiene buena **capcaidad de generalización**. Es **flexible mediante kernels** que le permiten modelar fronteras. Es **robusto ante el sobreajuste**, al maximizar el márgen y la regularización.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

SVM busca un hiperplano que maximice el margen entre las clases, los árboles de decisión dividen los datos en ramas según unas características.

Los Árboles de Decisión son más fáciles de interpretar y entender.

Los árboles de decisión pueden presentar un sobreajuste mayor frente al SVM.

SVM puede ser computacionalmente complejo, con grandes conjuntos de datos. 

SVM utiliza funciones de kernel para transformar los datos en un espacio de mayor dimensión. Los Árboles de Decisión no utilizan kernels.

1. ¿Cuáles son las principales características de un modelo de k vecinos cercanos (KNN)? 
2. ¿Cuáles son las principales diferencias entre un modelo de máquina de soporte vectorial y k vecinos cercanos?

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 02</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

Características principales:

No paramétrico: K-NN no asume una distribución previa (sobre los datos), útil cuando la relación del atributo no es comprendida.
Fácil de implementar: es facil de implementar y comprender, pero puede ser computacionalmente más costoso.
Sensible a la escala de las características: es necesario estandarizar los datos antes de aplicar KNN.
Sensible al ruido: es un algoritmo que depende de los datos de entrenamiento más cercanos. El ruido y los valores atípicos pueden afectar al rendimiento del modelo de manera negativa.
Puede ser lento para grandes conjuntos de datos: dado al calculo que hace, puede resultar lento para grandes cantidades de datos.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

Máquina de soporte vectorial vs K vecinos cercanos
* <u>Máquina de soporte vectorial</u>:

  **Ventajas**
  * Se pueden modelar realciones complejas, no lineales
  * Robusto al ruido, esto se debe a que maximizan los márgenes

  **Desventajas**
  * Necesidad de seleccionar una buena función kernel
  * Los parámetros del modelo son dificiles de interpretar
  * Requiere memoria significativa y poder de procesamiento

  **Utilidad**
  * Clasificación de texto e imágenes
  * Reconocimiento de escritura a mano



* <u>K vecinos cercanos</u>:

  **Ventajas**
  * Simple
  * Potente
  * Entrenamiento rápido
  * Puedo manejar naturalmente problemas extremos de multiclases, como etiquetadode texto

  **Desventajas**
  * Cosoto y lento para predecir nuevas instancias
  * Se debe definir ua función de distancia significativa
  * Funciona mal en conjuntos de datos de alta dimensionalidad

  **Utilidad**
  * Seguridad informática: detección de intrusos
  * Sistema de recomendación
  * Prlobmea de corrección ortográfica
  * Recuperación de contenido de video

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Decision tree:
* `cp` (parámetro de complejidad): controla el tamaño del arbol, donde valores más pequeños permiten árboles más complejos y valores más grandes conducen a árboles más pequeños.
* `rpart`: para el random forest, utiliza todas las variables consideradas
* `preProcess= c("center", "scale")`: para incorporar algunas de las tareas de limpieza y transformación

SVM:
* `c`: controla el equilibrio entre dos objetivos
* `tuneGrid`: define lso valores que se probarán para el parámetro `C` durante la búsqueda en la cuadrícula.

k vecinos cercanos:
* `k`: parámetro óptimo de vecinos cercanos en el que se minimiza el error de clasificación
* `scale = TRUE`: significa que las importaciones son escaladas para que el máximo sea *n* en la función `varImp()`

```r
set.seed(34)

# conjunto de entrenamiento y de test
index = caret::createDataPartition(datos$default, p = 0.8, list = FALSE)
```

```r
# Dentrenamiento y test
train_set = datos[index, ]
test_set  = datos[-index, ]
```

```r
# como están repartidos los datos en train
table(train_set$default) |> prop.table()
```

```r
# como están repartidos los datos en test
table(test_set$default) |> prop.table()
```

```r
# Configuramos el control de entrenamiento para usar validación cruzada de 10 pliegues, lo que ayuda a evitar
train1 = caret::trainControl(method = "repeatedcv", number = 10, repeats = 10)
```

```r
# Entrenamos el modelo utilizando el modelo de árbol de decisión (rpart)
modelo_dt = caret::train(default ~ income + balance + student , data = train_set, method = "rpart", trControl = train1)
```

```r
print(modelo_dt)
```

Este modelo tiene 8001 observaciones con 2 clases 0 y 1. El valor de cp para para la precisión fue 0.04119850, siendo la precisión un 0.9712419, esta es bastante alta casi 1.

```r
# Obtenemos información sobre el modelo Árbol de decisión seleccionado
dt_model = modelo_dt$finalModel; dt_model
```

```r
library(rpart.plot)
```
