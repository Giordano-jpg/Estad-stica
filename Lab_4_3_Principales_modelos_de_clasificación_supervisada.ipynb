Características principales:
* **No paramétrico:** K-NN no asume una distribución previa (sobre los datos), útil cuando la relación del atributo no es comprendida.
* **Fácil de implementar:** es facil de implementar y comprender, pero puede ser computacionalmente más costoso.
* **Sensible a la escala de las características:** es necesario estandarizar los datos antes de aplicar KNN.
* **Sensible al ruido:** es un algoritmo que depende de los datos de entrenamiento más cercanos. El ruido y los valores atípicos pueden afectar al rendimiento del modelo de manera negativa.
* **Puede ser lento para grandes conjuntos de datos:** dado al calculo que hace, puede resultar lento para grandes cantidades de datos.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

Máquina de soporte vectorial vs K vecinos cercanos
* <u>Máquina de soporte vectorial</u>:

  **Ventajas**
  * Se pueden modelar realciones complejas, no lineales
  * Robusto al ruido, esto se debe a que maximizan los márgenes

  **Desventajas**
  * Necesidad de seleccionar una buena función kernel
  * Los parámetros del modelo son dificiles de interpretar
  * Requiere memoria significativa y poder de procesamiento

  **Utilidad**
  * Clasificación de texto e imágenes
  * Reconocimiento de escritura a mano



* <u>K vecinos cercanos</u>:

  **Ventajas**
  * Simple
  * Potente
  * Entrenamiento rápido
  * Puedo manejar naturalmente problemas extremos de multiclases, como etiquetadode texto

  **Desventajas**
  * Cosoto y lento para predecir nuevas instancias
  * Se debe definir ua función de distancia significativa
  * Funciona mal en conjuntos de datos de alta dimensionalidad

  **Utilidad**
  * Seguridad informática: detección de intrusos
  * Sistema de recomendación
  * Prlobmea de corrección ortográfica
  * Recuperación de contenido de video

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Decision tree:
* `cp` (parámetro de complejidad): controla el tamaño del arbol, donde valores más pequeños permiten árboles más complejos y valores más grandes conducen a árboles más pequeños.
* `rpart`: para el random forest, utiliza todas las variables consideradas
* `preProcess= c("center", "scale")`: para incorporar algunas de las tareas de limpieza y transformación

SVM:
* `c`: controla el equilibrio entre dos objetivos
* `tuneGrid`: define lso valores que se probarán para el parámetro `C` durante la búsqueda en la cuadrícula.

k vecinos cercanos:
* `k`: parámetro óptimo de vecinos cercanos en el que se minimiza el error de clasificación
* `scale = TRUE`: significa que las importaciones son escaladas para que el máximo sea *n* en la función `varImp()`
