```r
# Eliminar la primera columna
datosn = select(BreastCancer, -Id)
```

```r
# Clases de la base de datos 'BreastCancer'
class(datosn)
```

```r
# Variables de la base de datos 'BreastCancer'
names(datosn)
```

```r
# Seleccionamos las 3 variables que nos interesan
zdata = datosn[, c("Cl.thickness", "Cell.size", "Cell.shape", "Class")]
```

```r
# Visualizamos la base de datos seleccionada
head(zdata)
```

```r
# Comprobamos si existen datos NAs para en caso afirmativo, eliminarlos y que no nos den problemas
wdata = na.omit(zdata)
```

```r
# Podemos comprobar si habían datos NA
nrow(zdata) # Base de datos inicial
nrow(wdata) # Base de datos tras el na.omit
```

```r
# Analisis exploratorio
summary(wdata)
```

```r
# Gráfico de tablas
boxplot(wdata, main="Diagrama de Caja de 'wdata'", col="lightblue")
```

**Interpretación:** Ninguna de las variables selccionadas tiene valores atípicos, en todas las variables excepto la variable dependiente "class", tienen un rango intercuartílico bastante similar, en el caso del tamaño de la celula y su forma, representan ambos un mismo rango intercuartílico.
